```python
## VISUALIZATION ##
```

```python
# 1.Calculating MEAN of release_year
mean_release_year=df['release_year'].mean()
# create a horizontal bar plot
plt.figure(figsize=(5,3))
plt.barh(['Mean Release Year'],[mean_release_year],color='skyblue')
plt.xlabel('Year')
plt.title('Netflix Release Year')
plt.show()
```

```python
#2. TV Shows Vs Movies on netflix-Bar chart #

# Count the number of TV shows and movies
type_counts = df['type'].value_counts()

# Create a bar chart
plt.figure(figsize=(4,2))
type_counts.plot(kind='bar', color=['lightgreen', 'salmon'])
plt.xlabel('Type')
plt.ylabel('Count')
plt.title('TV Shows vs Movies on Netflix')
plt.xticks(rotation=0)  # Rotate x labels if needed
plt.show()
```

```python
#3. Type of shows percentage on Netflix using piechart #
# Count the number of each type of show
type_counts = df['type'].value_counts()

# Create a pie chart
plt.figure(figsize=(4,4))
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', colors=['gold', 'blue'])
plt.title('Percentage of Show Types on Netflix')
plt.show()
```

```python
# 4.SHOW NETFLIX RATING DISTRIBUTION VISUALIZED BY COUNTPLOT #
# Create a count plot for the 'rating' column
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='rating', order=df['rating'].value_counts().index, palette='viridis')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.title('Distribution of Netflix Ratings')
plt.xticks(rotation=45)  # Rotate x-axis labels if needed
plt.show()
```

```python
#5. Show the Top 5 countries with Highest Movies/TV Shows, visualised by a PIE CHART.
# Count the number of shows per country
country_counts = df['country'].value_counts()

# Select the top 5 countries
top_5_countries = country_counts.head(5)

# Create a pie chart
plt.figure(figsize=(5,5))
plt.pie(top_5_countries, labels=top_5_countries.index, autopct='%1.1f%%', colors=['blue', 'salmon', 'lightgreen', 'gold', 'coral'])
plt.title('Top 5 Countries with Highest Number of Movies/TV Shows on Netflix')
plt.show()
```

```python
#6.Show the Top 5 Netflix country distribution separated by type of release (i.e Movies, TV shows), visualizing via Bar Graph.#
# Count the number of movies and TV shows per country
country_type_counts = df.groupby(['country', 'type']).size().reset_index(name='count')

# Get the top 5 countries with the highest number of shows/movies
top_countries = country_type_counts.groupby('country')['count'].sum().nlargest(5).index
filtered_data = country_type_counts[country_type_counts['country'].isin(top_countries)]

# Create a bar plot
plt.figure(figsize=(12, 8))
sns.barplot(data=filtered_data, x='country', y='count', hue='type', palette='viridis')
plt.xlabel('Country')
plt.ylabel('Count')
plt.title('Top 5 Countries by Type of Release on Netflix')
plt.xticks(rotation=45)  # Rotate x-axis labels if needed
plt.legend(title='Type')
plt.show()
```

```python
#7.Show the top 10 genres of movies and also tv shows.
df['genre'] = df['genre'].str.split(',')
df = df.explode('genre')

# Filter data for movies and TV shows
movies_df = df[df['type'] == 'Movie']
tv_shows_df = df[df['type'] == 'TV Show']

# Get top 10 genres for movies
top_10_movie_genres = movies_df['genre'].value_counts().head(10)

# Get top 10 genres for TV shows
top_10_tv_genres = tv_shows_df['genre'].value_counts().head(10)

# Plotting
fig, axes = plt.subplots(2, 1, figsize=(12, 12))

# Top 10 movie genres
top_10_movie_genres.plot(kind='bar', ax=axes[0], color='gold')
axes[0].set_title('Top 10 Movie Genres on Netflix')
axes[0].set_xlabel('Genre')
axes[0].set_ylabel('Count')
axes[0].set_xticklabels(top_10_movie_genres.index, rotation=45)

# Top 10 TV show genres
top_10_tv_genres.plot(kind='bar', ax=axes[1], color='coral')
axes[1].set_title('Top 10 TV Show Genres on Netflix')
axes[1].set_xlabel('Genre')
axes[1].set_ylabel('Count')
axes[1].set_xticklabels(top_10_tv_genres.index, rotation=45)

plt.tight_layout()
plt.show()
```

```python
#8.When is the Best to time to release a Movie/TV shows? Show it using heatmap

# Ensure 'Date Added' is in datetime format
#df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# Extract month and day of the week
df['Month'] = df['date_added'].dt.month
df['DayOfWeek'] = df['date_added'].dt.day_name()

# Count the number of releases by month and day of the week
release_counts = df.groupby(['Month', 'DayOfWeek']).size().unstack()

# Display the aggregated data
print("Aggregated release counts by month and day of the week:")
print(release_counts)

# Create a heatmap of releases by month and day of the week
plt.figure(figsize=(12, 8))
sns.heatmap(release_counts, annot=True, cmap="YlGnBu", fmt="d", cbar=True, linewidths=0.5)

# Customize the plot
plt.title('Number of Releases by Month and Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Month')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.show()
```
